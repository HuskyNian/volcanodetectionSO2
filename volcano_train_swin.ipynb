```python
import cv2
import json
from pycocotools.coco import COCO
annotation_file_path = '/content/drive/MyDrive/Volcano/random_split_folders/split1/test/test_coco_random1.json'
# Initialize COCO api for instance annotations
coco=COCO(annotation_file_path)
image_dir = '/content/drive/MyDrive/Volcano/random_split_folders/split1/test/images'
# Load and display a sample image
img_ids = coco.getImgIds()
img_data = coco.loadImgs(img_ids[0])[0]
image = cv2.imread('%s/%s'%(image_dir, img_data['file_name']))

# Load and display instance annotations
ann_ids = coco.getAnnIds(imgIds=img_data['id'])
anns = coco.loadAnns(ann_ids)
for ann in anns:
    x, y, w, h = ann['bbox']
    print('bbox',x,y,w,h)
    cv2.rectangle(image, (x, y), (x+w, y+h), (255,0,0), 2)
    cv2.putText(image, "plume", (x, y), cv2.FONT_HERSHEY_SIMPLEX , 0.5, (255,255,255), 2, cv2.LINE_AA)

cv2.imshow('Ground Truth', image)
cv2.waitKey(0)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
from pycocotools.coco import COCO

annotation_file_path = '/content/drive/MyDrive/Volcano/random_split_folders/split1/test/test_coco_random1.json'
# Initialize COCO api for instance annotations
coco=COCO(annotation_file_path)
image_dir = '/content/drive/MyDrive/Volcano/random_split_folders/split1/test/images'
# Load and display a sample image
img_ids = coco.getImgIds()
img_data = coco.loadImgs(img_ids[10])[0]
image = Image.open(f"{image_dir}/{img_data['file_name']}")

# Create figure and axes
fig, ax = plt.subplots(1)

# Display the image
ax.imshow(image)

# Load and display instance annotations
ann_ids = coco.getAnnIds(imgIds=img_data['id'])
anns = coco.loadAnns(ann_ids)
for ann in anns:
    x, y, w, h = ann['bbox']
    print('bbox',x, y, w, h)
    rect = patches.Rectangle((x, y), w, h, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    plt.text(x, y, "plume", bbox=dict(fill=False, edgecolor='red', linewidth=2))

plt.show()
```
